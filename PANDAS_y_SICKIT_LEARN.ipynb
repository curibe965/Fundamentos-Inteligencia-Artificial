```python
#Normalización de los datos
min_vals = df.min()
max_vals = df.max()

#Aplicar la normalización de min-max
df = (df - min_vals) / (max_vals - min_vals)
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.title('Años de experiencia vs Salario asociado normalizados')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
x = df.iloc[:,0].values
y = df.iloc[:,1].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

image.png


```python
#Esta función viene representado por la image de arriba lo que buscamos es
#actuaizar los pesos W y B
#Esta función viene representado por la image de arriba lo que buscamos es
#actuaizar los pesos W y B
def regresion_lineal_mco(x_train,y_train,x_test,y_test):

  x_mean = np.mean(x_train)
  y_mean = np.mean(y_train)

  n = len(x_train)

  numerador = 0
  denominador = 0

  for i in range(n):

    numerador += (x_train[i] - x_mean) * (y_train[i] - y_mean)
    denominador += (x_train[i] - x_mean)**2

  w = numerador / denominador
  b = y_mean - (w * x_mean)

  y_modelo = x_train * w + b
  plt.plot(x_train,y_train,'o')
  plt.plot(x_train, y_modelo)
  plt.title('Años de experiencia vs Salario asociado normalizados entrenamiento')
  plt.xlabel('Años de experiencia')
  plt.ylabel('Salario')
  plt.show()

  y_modelo = x_test * w + b
  plt.plot(x_test,y_test,'o')
  plt.plot(x_test, y_modelo)
  plt.title('Años de experiencia vs Salario asociado normalizados entrenamiento')
  plt.xlabel('Años de experiencia')
  plt.ylabel('Salario')
  plt.show()

  Error = (y_train - (x_train * w + b))
  MSE = np.sum(Error**2)/n
  print('El MSE para los datos de entrenamiento es: '+ str(MSE ))

  Error = (y_test - (x_test * w + b))
  MSE = np.sum(Error**2)/n
  print('El MSE para los datos de testeo es: '+ str(MSE ))
```

```python
regresion_lineal_mco(x_train,y_train,x_test,y_test)
```

```python
#En este fragmento de cogdigo buscamos realizar el regresor lineal por medio de
#la función LinearRegression(), lo entrenamos con .fit(),
#Es necesario crear la variale x_trains_rs debido a que el regresor necesita
#tener una entrada de dimensiones (n,1), verifique las dimensiones de
#x_train.shape y las de x_train_rs.shape

x_train_rs = np.reshape(x_train,(len(x_train),1))
y_train_rs = np.reshape(y_train,(len(y_train),1))
linear_model = LinearRegression().fit(x_train_rs, y_train_rs)
```

image.png


```python
#La función .score retorna el error R2
print('Punetaje Entrenamientos: ', linear_model.score(x_train_rs, y_train_rs))
```

```python
y_modelo_train = linear_model.predict(x_train_rs)
plt.plot(x_train,y_train,'o')
plt.plot(x_train, y_modelo_train)
plt.title('Años de experiencia vs Salario asociado normalizados entrenamiento')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()


x_test_rs = np.reshape(x_test,(len(x_test),1))
y_modelo_test = linear_model.predict(x_test_rs)
plt.plot(x_test,y_test,'o')
plt.plot(x_test, y_modelo_test)
plt.title('Años de experiencia vs Salario asociado normalizados testeo')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()
```

```python
#Importar métricas
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test, y_modelo_test)
print(mse)
```
